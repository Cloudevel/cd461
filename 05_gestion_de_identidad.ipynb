# *GitHub Enterprise*.

## Opciones de *GitHub* para organizaciones.

Además de opción gratuita para organizaciones, *GitHub* ofrece dos opciones de suscripción pagada para las organizaciones.

* [*GitHub Team*](https://docs.github.com/en/get-started/learning-about-github/githubs-products#github-team), para equipos de trabajo.
* [*GitHub Enterprise*](https://docs.github.com/en/get-started/learning-about-github/githubs-products#github-enterprise), para empresas.

Ambas opciones ofrecen funcionalidades adicionales como:

* Administración general de *GitHub*.
* Gestión de identidad y autenticación.
* Funcionalidades completas para repositorios privados.
* Seguridad avanzada.

Los precios son calculados con base en el número de usuarios de la organización.

### Opciones de *GitHub Enterprise*.

* [*GitHub Enterprise Cloud*](https://docs.github.com/es/enterprise-cloud@latest/admin/overview/about-github-enterprise-cloud), el cual es un servicio alojado en la nube en la modalida de  *SaaS*. Esta será la opción que se explorará en este curso.
* [*GitHub Enterpise Server*](https://docs.github.com/es/enterprise-server@3.5/admin/overview/about-github-enterprise-server), el cual permite desplegar el servicio alojado en las instalaciones *on premise* de la empresa. 

**Nota:** A partir de este punto y en los siguientes capítulos, al hacer mención de *Github Enterprise* se hará referencia a *Github Enterprise Cloud*.

### *URL* de las cuentas *Enterpise*.

Para acceder al servicio de *GitHub Enterpise*, se utiliza la siguiente estructura de *URL*.

```
github.com/enterprises/<identificador>
```



## Cuentas de *Github Enterprise*.

Una cuenta empresarial de *GitHub* es capaz de administrar múltiples organizaciones. Sin embargo, siempre debe de existir un gestor principal, el cual corrponde a la identidad de una cuenta de usuario en *GitHub* o de una organización.

En la configuración *Enterprise*, los propietarios de empresas pueden invitar a las organizaciones existentes a unirse a dicha cuenta, transferir organizaciones entre cuentas *Enterprise*, o crear nuevas organizaciones. 

Para obtener más información consultar:

* https://docs.github.com/es/enterprise-cloud@latest/admin/user-management/managing-organizations-in-your-enterprise/adding-organizations-to-your-enterprise

## Roles de *GitHub Enterprise*.

Es posible definir los siguientes roles a los miembros o grupos de miembros de una empresa.

* [Propietarios](https://docs.github.com/es/enterprise-cloud@latest/admin/user-management/managing-users-in-your-enterprise/roles-in-an-enterprise#enterprise-owners).
* [Miembros](https://docs.github.com/es/enterprise-cloud@latest/organizations/managing-peoples-access-to-your-organization-with-roles/roles-in-an-organization)
* [Moderadores](https://docs.github.com/es/enterprise-cloud@latest/organizations/managing-peoples-access-to-your-organization-with-roles/roles-in-an-organization#organization-moderators)
* [Gerentes de facturación](https://docs.github.com/es/enterprise-cloud@latest/organizations/managing-peoples-access-to-your-organization-with-roles/roles-in-an-organization#billing-managers)

## Equipos.

En *GitHub Enterprise*, cada miembro de la organización puede ser agregar a uno o varios equipos. 

Es posible crear equipos con permisos de acceso "en cascada". Los equipos facilitan la asignación de los permisos de los repositorios en un nivel más granular y habilitar la comunicación y notificación entre los miembros del equipo.

*GitHub* permite sincronizar los equipos con grupos de proveedores de identidades (IAM), como es el caso de *Azure Active Directory*. Al sincronizar un equipo de GitHub con Azure AD, los puede replicar cambios en GitHub automáticamente, lo que reduce la necesidad de actualizaciones manuales y scripts personalizados. Para consultar la lista de proveedores de indentidad compatibles, consultar la siguiente liga:

* https://docs.github.com/es/enterprise-cloud@latest/organizations/managing-saml-single-sign-on-for-your-organization/about-identity-and-access-management-with-saml-single-sign-on#supported-saml-services
